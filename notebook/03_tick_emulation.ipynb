```python
file_name = '../data/raw/ticks/usdjpy-tick-2020-01-01-2024-12-31.csv'
f = open(file_name)
f.readline()
```

```python
from datetime import datetime
import threading
import queue
import time
```

```python
ds = queue.Queue()
```

```python
def get_tick():
    tick = {}
    values = f.readline().rstrip("\n").split(",")
    ts = datetime.fromtimestamp(int(values[0])/1000)
    tick[ts] = float(values[1])
    return tick
```

```python
def emulate_tick_stream():
    while True:
        time.sleep(1)
        temp = get_tick()
        ds.put(temp)
```

```python
def trading_algo():
    while True:
        temp = ds.get()
        print('Received tick', temp)
```

```python
data_source_thread = threading.Thread(target = emulate_tick_stream)
data_receiver_thread = threading.Thread(target = trading_algo)
```

```python
data_source_thread.start()
data_receiver_thread.start()
```

```python
class SlidingWindow:
    def __init__(self, length):
        self.data = [0] * length

    def add(self, element):
        self.data.append(element)
        self.data.pop(0)
```

```python
sw = SlidingWindow(60)
```
